**FASTA FILES:** FASTA format is a text-based format for representing either nucleotide sequences or amino acid (protein) sequences, in which nucleotides or amino acids are represented using single-letter codes.Each sequence will have a “>” symbol in the beginning and it is restricted to that location only.

**FASTQ FILES:** A FASTQ file is a text file that contains the sequence data from the clusters that pass filter on a flow cell (for more information on clusters passing filter. Each FASTQ file consists of reads, each read in turn consisting of 4 lines.Every read will have “@” symbol in the beginning.However that is not the only position of the “@” symbol in the entire file.
